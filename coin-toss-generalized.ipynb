```python
import sympy as sp
```

```python
def create_vector(size: int):
    arr = [1]
    for i in range(0, size - 1):
        arr.append(0)
    return sp.Matrix(arr)
```

```python
def create_matrix(size: int):
    arr_arr = []
    for i in range(0, size):
        arr = []
        for j in range(0, size):
            if (i == 0 and j < size - 1) or (i - j == 1):
                arr.append(0.5)
            elif (i == size - 1) and (j == size - 1):
                arr.append(1)
            else:
                arr.append(0)
        arr_arr.append(arr)
    mat = sp.Matrix(arr_arr)
    return mat
```

```python
for s in range(1, 11):
    print(f"\nn = {s}")
    sp.pprint(create_matrix(s) ** 99 * create_vector(s))
```
